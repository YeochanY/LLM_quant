```python
import pandas as pd

data_1 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/bert/fin_expert_report_until_2010.csv')
data_2 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/bert/fin_expert_report_2010.csv', encoding='cp949')

```

```python
data_combined = pd.concat([data_1, data_2], ignore_index=True)
data_combined
```

```python
data_combined.sort_values(by=['작성일'], inplace=True)
data_combined
```

```python
!pip install transformers
from transformers import pipeline
```

```python

sentiment_classification = pipeline("text-classification", "snunlp/KR-FinBert-SC")
```

```python
sentiment_classification("today is payday!")
```

```python
sentiment_results = []
for index, row in data_combined.iterrows():
  result = sentiment_classification(row['제목'])
  sentiment_results.append({
      'label': result[0]['label'],
      'score': result[0]['score']
  })

data_combined['Sentiment Label'] = [result['label'] for result in sentiment_results]
data_combined['Sentiment Score'] = [result['score'] for result in sentiment_results]

data_combined
```

```python

data_combined.to_csv('/content/gdrive/MyDrive/Colab Notebooks/bert/data_combined_label.csv')
```

```python
def covert_to_df(temp):
  temp.columns = temp.iloc[7,:]   # columns명 A코드로 할당
  temp = temp.iloc[13:,:]         # 불필요한 행 삭제
  temp.index = temp.iloc[:, 0]    # index에 날짜 할당
  temp.index.name = 'date'
  temp = temp.drop(["Symbol"], axis = 1) # 중복되는 날짜 행 삭제
  return temp
```
