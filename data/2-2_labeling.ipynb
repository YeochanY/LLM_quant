## 데이터 임포트

```python
import pandas as pd

data_1 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/bert/리포트_데이터_~2010.csv')
data_2 = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/bert/리포트_데이터_2010.csv', encoding='cp949')

```

```python
data_combined = pd.concat([data_1, data_2], ignore_index=True)
data_combined
```

```python
data_combined['종목코드'] = data_combined['종목코드'].apply(lambda x: 'A' + str(x).zfill(6))
data_combined
```

```python

import re

def split_authors(authors):
  """
  Splits the author string into a list of authors.
  """
  return re.split('[,.]', authors)

data_combined['작성자1'] = data_combined['작성자'].apply(lambda x: split_authors(x)[0])
data_combined['작성자2'] = data_combined['작성자'].apply(lambda x: split_authors(x)[1] if len(split_authors(x)) > 1 else None)

data_combined
```

```python
close = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/bert/close price.csv", index_col=0)
close.head()
```

```python
import numpy as np
```

## 라벨링

```python
import datetime as dt

def find_future_price(row, day):
  date = dt.datetime.strptime(row['작성일'], '%Y-%m-%d')
  future_date = date + dt.timedelta(days=day)
  try:
    return close.loc[future_date.strftime('%Y-%m-%d'), row['종목코드']]
  except KeyError:
    # 종목코드가 close에 없는 경우 NaN 반환
    return np.nan

data_combined['today'] = data_combined.apply(find_future_price, day = 0, axis=1)
data_combined['After 3 months'] = data_combined.apply(find_future_price, day = 90, axis=1)
data_combined.head()
```

```python
data_combined['평가의견'].unique()
```

```python
def change_label(row):
    positive_labels = ['매수', 'BUY', 'STRONGBUY', '적극매수', 'TRADINGBUY', 'StrongBuy', 'TradingBuy', 'Buy', 'buy',  'OUTPERFORM', 'Outperform', '비중확대']
    negative_labels = ['매도', 'SELL', '비중축소', 'Reduce', 'Sell', 'underperform', 'UNDERPERFORM']
    neutral_labels = ['중립', 'HOLD', '보유', '시장평균', 'MARKETPERFORM', 'MKTPER', 'MKTPERFORM', 'mktper', 'MKTPERF', 'NEUTRAL', 'Neutral', 'neutral', 'marketperform', '시장수익률', 'HOLD', 'Hold', 'NotRated', 'nr', 'na', 'nan', '투자의견없음', '-', 'SUSPENDED']

    if row['평가의견'] in positive_labels:
        #return 'POS'
        return 1
    elif row['평가의견'] in negative_labels:
        #return 'NEG'
        return -1
    elif row['평가의견'] in neutral_labels:
        #return 'NEU'
        return 0

    else:
        #return 'UNRATED'  # For any unspecified or unrecognized labels
        return 0

# Applying the updated function
data_combined['label 1'] = data_combined.apply(change_label, axis=1)
data_combined.head()
```

```python
import numpy as np

def compare_prices(row):
  if row['After 3 months'] > row['today']:
    return 1
  elif row['After 3 months'] < row['today']:
    return -1
  else:
    return 0  # Prices are equal

data_combined['pred'] = data_combined.apply(compare_prices, axis=1)
data_combined.head()
```

```python
from sklearn.metrics import accuracy_score, f1_score
```

```python
data_combined['label 1'].unique()
```

```python

# Calculate accuracy
accuracy = accuracy_score(data_combined['label 1'], data_combined['pred'])
f1 = f1_score(data_combined['label 1'], data_combined['pred'], average='weighted')

# Print the results
print("Accuracy:", accuracy)
print("F1 Score:", f1)
```

```python
data_combined['true_pred'] = (data_combined['label 1'] == data_combined['pred']) * 1
```

```python
mean_true_pred_by_agency = data_combined.groupby('작성기관')['true_pred'].mean()
mean_true_pred_by_agency
```

```python
def replace_agency_names(df, replacements):
  for old_name, new_name in replacements.items():
    df['작성기관'] = df['작성기관'].replace(old_name, new_name)
  return df


replacements = {
    '신한금융': '신한금융투자',
    '신한증권': '신한금융투자',
    '한투증권': '한국투자증권',
    '한화증권': '한화투자증권',
    '유진증권': '유진투자증권',
    '하이증권': '하이투자증권',
    '미래에셋대우': '미래에셋증권',
    'NH농협증권': 'NH투자증권',
    'NH증권': 'NH투자증권'


}

data_combined = replace_agency_names(data_combined.copy(), replacements)
print(data_combined['작성기관'].unique())
```

```python
mean_true_pred_by_agency = data_combined.groupby('작성기관')['true_pred'].mean()
mean_true_pred_by_agency
```

## bert 모델과 비교

```python
bert = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/bert/data_combined_label.csv')
bert
```

```python
bert['Sentiment Label'] = bert['Sentiment Label'].apply(lambda x: 'POS' if x == 'positive' else 'NEG' if x == 'negative' else 'NEU' if x == 'neutral' else x)
```

```python
from sklearn.metrics import accuracy_score, f1_score
```

```python
data_combined_1 = data_combined[data_combined['label 1'] != 'UNRATED']['label 1']
bert_labels_1 = bert[bert.index.isin(data_combined_1.index)]['Sentiment Label']


# Calculate accuracy
accuracy = accuracy_score(data_combined_1, bert_labels_1)
f1 = f1_score(data_combined_1, bert_labels_1, average='weighted')

# Print the results
print("Accuracy:", accuracy)
print("F1 Score:", f1)

```

```python
from sklearn.metrics import classification_report

report = classification_report(data_combined_1, bert_labels_1)
print(report)
```

```python
pos_neg_index = bert[bert['Sentiment Label'].isin(['POS', 'NEG'])].index
data_combined_filtered = data_combined.loc[pos_neg_index]
```
